```python
def heatmap_stations(station_dict, val_f):
    heatmap_data_list = []
    for station in station_dict.values():
        val = val_f(station)
        #print val
        line = "{location: new google.maps.LatLng(%s, %s), weight: %0.1f}" % \
                (station['latitude'], station['longitude'], val)
        heatmap_data_list.append(line)
    heatmap_data = ',\n  '.join(heatmap_data_list)
    javascript_template = """
var map, pointarray, heatmap;

function hmap_initialize() {

    var heatMapData = [
      %s
    ];
    var newYork = new google.maps.LatLng(40.728,-73.99);
    
    map = new google.maps.Map(document.getElementById('map-canvas%s'), {
      center: newYork,
      zoom: 13,
      mapTypeId: google.maps.MapTypeId.SATELLITE
    });
    
    var heatmap = new google.maps.visualization.HeatmapLayer({
      data: heatMapData
    });
    heatmap.setMap(map);
}

function loadScript() {

  if (typeof google === 'object' && typeof google.maps === 'object') {
    console.log('Google initialized');
    
    hmap_initialize();
  } else {
    console.log('Google not initialized');
    var script = document.createElement("script");
    script.type = "text/javascript";
    script.src = "https://maps.googleapis.com/maps/api/js?sensor=false&libraries=visualization&callback=hmap_initialize";
    document.body.appendChild(script);
  }
    
        
}

setTimeout(loadScript, 1000);

"""
    id_ = ("%.3f" % rand())[2:]
    js = "<script>%s</script>" % (javascript_template % (heatmap_data, id_))
    return d.HTML('''<div id="map-canvas%s" style="width: 500px; height: 500px"></div> %s''' % (id_, js))
```

Heatmap functions
=================
I have created a heatmap_stations function that draws aggregations.  Here I show a heatmap of the most active stations from the past week

```python
heatmap_stations(stations_by_id, lambda x: x.get('starting',{}).get('week',0))
```

```python
#these calculations grab all weekday and weekend starting trips
ss.starting_trips['weekday'] = ss.starting_trips.index.map(lambda x: x.weekday())
weekday_starting_trips = ss.starting_trips[ss.starting_trips['weekday'] < 5]
weekend_starting_trips = ss.starting_trips[ss.starting_trips['weekday'] >= 5]
```

```python

heatmap_stations(stations_by_id, lambda x: weekday_starting_trips.get(int(x['id']), np.array([])).sum())
```

```python
heatmap_stations(stations_by_id, lambda x: weekend_starting_trips.get(int(x['id']), np.array([])).sum())
```
